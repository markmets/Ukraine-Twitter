# 02 — Preprocessing

Loads raw tweet CSV files from `../data/raw/`, cleans and deduplicates them,
and writes the cleaned dataset to `../data/processed/tweets_clean.csv`.

```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path('../data/raw')
PROCESSED_DIR = Path('../data/processed')
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# Load all raw CSV files
raw_files = sorted(RAW_DIR.glob('ukraine_tweets_*.csv'))
print(f'Found {len(raw_files)} raw files')

dfs = []
for f in raw_files:
    df = pd.read_csv(f, parse_dates=['created_at'])
    dfs.append(df)

if dfs:
    tweets = pd.concat(dfs, ignore_index=True)
    print(f'Total rows before cleaning: {len(tweets):,}')
else:
    print('No raw files found — place CSV files in data/raw/ first')
```
