# 01 — Data Collection

Collects tweets mentioning Ukraine from the Twitter API (v2) for the period 2008–2023
and exports them as CSV files to `../data/raw/`.

**Requirements:**
- A valid Twitter API v2 Bearer Token stored in `bearer_token.txt` (project root)
- `tweepy` installed (`pip install tweepy`)

```python
import os
import tweepy
import pandas as pd
from pathlib import Path

RAW_DATA_DIR = Path('../data/raw')
RAW_DATA_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# Load bearer token
bearer_token_path = Path('../bearer_token.txt')
if bearer_token_path.exists():
    BEARER_TOKEN = bearer_token_path.read_text().strip()
else:
    BEARER_TOKEN = os.environ.get('TWITTER_BEARER_TOKEN', '')

client = tweepy.Client(bearer_token=BEARER_TOKEN, wait_on_rate_limit=True)
```

```python
# Define search parameters
QUERY = '(ukraine OR #ukraine OR #Ukraine OR #Україна) lang:en -is:retweet'
TWEET_FIELDS = ['created_at', 'text', 'lang', 'public_metrics', 'author_id']

# TODO: iterate over date ranges 2008-2023 and paginate results
# Note: full-archive search requires Twitter API v2 Academic Research access
print('Search query:', QUERY)
```
